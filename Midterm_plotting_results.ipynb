Plot u'd for the 3 cases where the domain 0<=x<=L is divided into 4,8, and 16 intervals. Also plot the true derivative, u'.
What conclusions do you make?

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
num_points=16
L=1
x=np.linspace(0,L,num_points+1)
u_now=np.cos((2*np.pi*x/L))
```

```python
u_now
```

```python
plt.plot(x, u_now)
plt.xlabel('Space')
plt.ylabel(' U Velocity (m/s)')
plt.title('Initial Condition')
```

```python
#Plot both u'd and true derivative
```

```python
delta_x=0.05
```

```python
#finite differencing
u_finite=-(np.sin((2*np.pi*x)/L)*np.sin((2*np.pi*delta_x)/L))/(delta_x)
u_finite
```

```python
plt.plot(x, u_finite)
plt.xlabel('Space')
plt.ylabel(' U Velocity (m/s)')
plt.title('Centered difference derivative divided into 16 intervals')
```

```python
u_prime_real=-((2*np.pi)/L)*np.sin((2*np.pi*x)/L)
u_prime_real
```

```python
plt.plot(x, u_prime_real)
plt.xlabel('Space')
plt.ylabel(' U Velocity (m/s)')
plt.title('Actual derivative divided into 16 intervals')
```

```python
plt.plot(x, u_prime_real,label='Actual derivative')
plt.plot(x,u_finite,label='Centered difference approximation')
plt.xlabel('Space')
plt.ylabel(' U Velocity (m/s)')
plt.legend()
plt.title('Comparion of results for delta_x=0.05, divided into 16 intervals')
```

The actual derivative here is approximated quite nicely by the centered differencing with 16 intervals

```python
ratio=u_finite/u_prime_real
ratio
```

The ratio is nearly 1 everywhere (0.935) for delta_x =0.1 and 0.983 for delta_x=0.05. As delta_x approaches 0, the ratio converges to 1.
