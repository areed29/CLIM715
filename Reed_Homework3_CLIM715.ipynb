### CASE 1: Use forward differences

The frequency is 2pi/T where period T=ndelta_T and n is an integer. Thus, you can use wdeltaT=2pi/n

Choose a variety of values of w (ie. of n), to show that this scheme is unstable.

```python
time_steps=20 # value of n (w) chosen 
```

```python
u_amplitudes=[]
u_initial=1
for n in range(1,time_steps+1):
    u_next=u_initial*(1+(1j*2*np.pi)/n) 
    #j here is used to represent a complex number in python, so 'i' in the initial equation is j here.
    u_initial=u_next
    u_amplitudes.append(u_next)
```

```python
amplitudes=np.asarray(u_amplitudes)
```

```python
time_vals=np.arange(0,20,1)
```

```python
#print(amplitudes)
```

```python
plt.plot(time_vals,amplitudes)
plt.xlabel('Time step')
plt.ylabel('Amplitude')
plt.axhline(1,linestyle='dashed',color='red')
plt.title('Integrated Oscillation equation using forward differencing')
```

Analysis: The amplitude grows larger at each timestep so the solution is unstable

### CASE 2: Use backward differences

```python
time_steps=20
```

```python
u_amplitudes = []
u_initial = 1

for n in range(1,time_steps+1):
    u_next = u_initial / (1 - 1j * (2 * np.pi) / (n))
    u_initial = u_next
    u_amplitudes.append(u_next)
```

```python
amplitudes=np.asarray(u_amplitudes)
```

```python
plt.plot(time_vals,amplitudes)
plt.xlabel('Time step')
plt.ylabel('Amplitude')
plt.axhline(1,linestyle='dashed',color='red')
plt.title('Integrated Oscillation equation using backward differencing')
```

Analysis: the solution is less than 1, but hovers around 0 for each time step, so the solution is stable.

### CASE 3: Use the trapezoidal scheme

```python
time_steps=20
```

```python
u_amplitudes=[]
u_initial=1
for n in range(1,time_steps+1):
    u_next=u_initial*(1 +1j * (2 * np.pi) / (n)*2)/ (1 - 1j * (2 * np.pi) / (n)*2) #need u_next on LHS of equation..
    # u_next is solved for in terms of u_initial as in problem above.
    u_initial=u_next
    u_amplitudes.append(u_next)
```

Analysis: The solution is less than 1, but hovers on either side of 0 and oscillates. It does not return to its initail state but does not grow unbounded, therefore it is neutrally stable.

### CASE 4: Use the Matsuno scheme
