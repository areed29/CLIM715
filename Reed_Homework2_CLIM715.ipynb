Integrate the linear first order constant wind speed advection equation for the following cases:

du/dt +c du/dx =0

### Case #1

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ntotal=64 #final time step
delta_t=1/64 #time interval of 30 seconds
c=1
delta_x=0.0625 #grid spacing
nx =16 #number of grid points
mu=0.25 #CFL condition, c*delta_t/delta_x
```

```python
#First, plot initial condition
num_points=16
xmax=1
x=np.linspace(0,xmax,num_points+1)
u_now=np.sin(2*np.pi*x)
```

```python
plt.plot(x, u_now)
plt.xlabel('Space')
plt.ylabel(' U Velocity (m/s)')
plt.title('Initial Condition')
```

```python
#time loop for centered scheme
#u_next=np.zeros_like(x)
```

```python
for n in range(0,ntotal):
    u_next=np.copy(u_now)
    u_past=np.roll(u_now,1)
    for i in range(1, nx):
        # Compute spatial differences using central differences
        d_q_dx = (u_now[i + 1] - u_now[i - 1]) / (2*delta_x)
        # Update q for the next time step
        u_next[i] = u_past[i] - delta_t * (c * d_q_dx)
```

```python
### Apply boundary conditions
u_next[0]=u_now[0]
u_next[nx-1]=u_now[nx-1]
```

```python
plt.plot(x, u_next)
plt.xlabel('Space')
plt.ylabel(' U Velocity (m/s)')
plt.title('Integrated solution for mu=0.25')
```

Observation: soltion propagates eastward but amplitude does not exceed 1.0
